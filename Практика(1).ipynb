## 3. Проверьте корректность наименований столбцов, соответствие «змеиному регистру». Приведите наименования столбцов и весь текст в датафрейме к единому (нижнему) регистру.

```python
def to_snake(name: str) -> str: #Переводит строку из CamelCase или с пробелами в snake_case.
    result = []
    n = len(name)
    for i, ch in enumerate(name):
        # заменяем пробел на подчеркивание
        if ch == ' ':
            result.append('_')
            continue

        # определяем, надо ли вставить '_'
        if ch.isupper() and i > 0:
            prev = name[i-1]
            nxt = name[i+1] if i+1 < n else None

            # 1) строчная граница →Прописная
            # 2) цифра → Прописные буквы
            # 3) аббревиатуры конца: верхний регистр → верхний регистр, затем нижний регистр.

            if prev.islower() or prev.isdigit() or (prev.isupper() and nxt and nxt.islower()):
                result.append('_')

        result.append(ch.lower())

    return ''.join(result)
```

```python
original_cols = df.columns.tolist()

new_cols = [to_snake(col) for col in original_cols]

df.columns = new_cols
print("Столбцы переименованы по 'змеиному регистру':\n")
df.columns.tolist()
```

Приводим все текстовые данные (object) к нижнему регистру и убираем лишние пробелы.

```python
for col in df.select_dtypes(include=['object']):
    df[col] = df[col].str.lower().str.strip()
```

## 4. Проверьте датафрейм на пропуски. Заполните пропущенные значения. Для каждой замены укажите, почему пропущенные значения заполняются именно таким образом. (NB! Если строки с пропусками составляют не более 5% от общего числа строк и заполнить их не представляется возможным, в этом случае строки можно удалить).

Сначала вычислим долю пропусков для каждого столбца (%):

```python
missing_pct = df.isna().mean() * 100
missing_pct
```

Логика заполнения или удаления:
- Если пропусков <=5%, и нет возможности заполнить осмысленно — удаляем строки
- Иначе заполняем средними/медианными значениями (для числовых) или наиболее частыми (для категориальных)

```python
threshold = 5.0
for col in df.columns:
    pct = missing_pct[col]
    if pct == 0:
        continue
    if pct <= threshold:
        df = df[df[col].notna()]
        print(f"Строки с отсутствующим {col} удалены (<= {threshold}% отсутствующих)")
    else:
        # Заполняем в зависимости от типа столбца
        if pd.api.types.is_numeric_dtype(df[col]): # Если числовой тип
            median_val = df[col].median()
            df[col].fillna(median_val, inplace=True)
            print(f"Отсутствует числовой столбец '{col}', заполненный медианой = {median_val}")
        else: # Если категориальный тип
            mode_val = df[col].mode()[0]
            df[col].fillna(mode_val, inplace=True)
            print(f"Отсутствует категориальный столбец '{col}', заполненный режимом = '{mode_val}'")
```

## 5. Измените типы данных в столбцах, где это необходимо, объясните изменения.

```python
def parse_dates(df: pd.DataFrame, date_cols: list = None) -> pd.DataFrame:
    df_parsed = df.copy()

    if date_cols is None:
        date_cols = [c for c in df_parsed.columns if 'date' in c.lower()]

    for col in date_cols:
        try:
            df_parsed[col] = pd.to_datetime(
                df_parsed[col],
                dayfirst=False,
                errors='coerce',
                infer_datetime_format=True
            )
            print(f"Колонка '{col}' сконвертирована в datetime.")

        except Exception as e:
            print(f"Не удалось сконвертировать колонку '{col}': {e}")

    return df_parsed


df = parse_dates(df) # Если есть колонка с датами в строковом формате — переводим её в datetime

# Также можно понизить типы целочисленных столбцов для экономии памяти
if df['customer_id'].dtype == 'int64':
    df['customer_id'] = df['customer_id'].astype('int32') # Меняем кодировк на более низкую
```

##4. В рамках бизнес-задачи постройте несколько сводных таблиц (не менее трех). Объясните полученные результаты.

Добавим столбцы общих покупок (онлайн и оффлайн) и величины скидки, далее будем неоднократно использовать

```python
df['total_spend'] = df['offline_spend'] + df['online_spend']
df['discount_amount'] = df['total_spend'] * df['discount_pct'] / 100
```

###Вклад каналов в общие продажи по месяцам (онлайн и оффлайн)

```python
pivot_month = pd.pivot_table(
    df,
    index='month',
    values=['offline_spend', 'online_spend'],
    aggfunc='sum'
).sort_index()

pivot_month['total_spend'] = pivot_month['offline_spend'] + pivot_month['online_spend']
pivot_month['offline_spend_pct(%)'] = (pivot_month['offline_spend'] / pivot_month['total_spend']) * 100
pivot_month['online_spend_pct(%)'] = (pivot_month['online_spend'] / pivot_month['total_spend']) * 100

pivot_month
```

Что показывает и зачем нужно:
- Пик покупок офлайн приходится на декабрь, а онлайн - на август, но в сумме покупок больше в декабре.
- Офлайн продажи берут на себя больше, чем онлайн.

###Объем продаж и сумма скидок по товарным категориям

```python
pivot_category = pd.pivot_table(
    df,
    index='product_category',
    values=['total_spend', 'discount_amount'],
    aggfunc='sum'
).sort_values('total_spend', ascending=False)

pivot_category['discount_pct(%)'] = (pivot_category['discount_amount'] / pivot_category['total_spend']) * 100

pivot_category
```

Что показывает и зачем нужно:

- Мы видим, на сколько продано товаров из каждой категории, сколько было потрчено на скидки для товаров каждой категории.
- Позволяет понять, что товары из категории apparel приносят больше всего выручки, и стоит направлять маркетинг на них.


###Сегментация клиентов по локации, стажу и средней корзине

```python
def tenure_bucket(m): # Добавляем ярлыки в зависимости от "возраста"
    if m <= 6:    return 'New (≤6m)'
    if m <= 24:   return 'Regular (7–24m)'
    return 'Loyal (>24m)'

cust = df.groupby(['customer_id', 'location']).agg(
    tenure_months=('tenure_months', 'first'),
    avg_basket=('total_spend', 'mean')
).reset_index()

cust['tenure_segment'] = cust['tenure_months'].apply(tenure_bucket)

pivot_segment = pd.pivot_table(
    cust,
    index=['location', 'tenure_segment'],
    values=['customer_id', 'avg_basket'],
    aggfunc={'customer_id':'count', 'avg_basket':'mean'}
).rename(columns={'customer_id':'customer_count'})

pivot_segment
```

Что показывает и зачем нужно:
- В регионе Chicago больше всего лояльных покупателей, но самые большие средние корзины в регионе California, а регион Washington DC необходимо развивать.
- Помогает таргетировать акции: необходимо увеличивать продажи в регионе Washington DC и стимулировать лояльных покупателей в регионе California.
